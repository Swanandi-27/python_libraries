```python
#series is one dimensional labelled array , capable of holding any data type , the axis labels are collectively called the index
import numpy as np
import pandas as pd
```

```python
labels=['a','b','c']
my_list=[10,20,30]
my_arr=np.array([10,20,30])
my_dict={1:10,2:20,3:30}
d={'a':10,'b':20,'c':30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list,index=labels)
```

```python
pd.Series(my_arr)
```

```python
pd.Series(my_dict)
```

```python
pd.Series(d)
```
