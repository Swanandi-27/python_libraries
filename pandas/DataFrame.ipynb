```python
#DataFrame when you combine multiple series together
import numpy as np
import pandas as pd
```

```python
#create data frame using dictionary
data={
    'Name':['john', 'anna','peter','linda'],
    'Age':[28,34,29,42],
    'city':['new york','paris','berlin','london'],
    'Salary':[65000,70000,80000,62000]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
#create data frame using list 
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
col = ["Name","Age","City","Salary"]
df2 = pd.DataFrame(data_list,columns =col)
df2
```

```python
#selection and indexing of columns 
df2["Name"]
```

```python
#column selection
df2[["City","Name"]]
```

```python
#creating new column
df2["Designation"]=["Doctor","engineer","HR","lawyer"]
df2
```

```python
#Removing column from data
df2.drop('Designation',axis = 1,inplace=True)
df2
```

```python
#Removing row from data
df2.drop(0,axis = 0)
```
