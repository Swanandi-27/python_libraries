```python
import numpy as np
import pandas as pd
```

Finding Missing Data

```python
data = {
    'A':[1,2,np.nan,4,5],
    'B':[np.nan,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5]
}
df=pd.DataFrame(data)
df
```

```python
#isna helps to find null values
df.isna()
```

```python
#count the total null values 
df.isna().sum()
```

```python
#chcek if is their any null value in each column
df.isna().any()
```

Removing Null values

```python
#whole dataset is get delete
df.dropna()
```

```python
#threshold
df.dropna(thresh=3)  #remove the row only if it contains less than 3 non null values
```

Filling Missing Data

```python
#fill all the missing values with 0
df.fillna(0)
```

```python
#fill missing values with diff values for each column
values= {'A':0,'B':7,'C':2,'D':4}
df.fillna(value=values)
```
